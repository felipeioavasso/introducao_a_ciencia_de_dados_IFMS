A biblioteca Matplotlib é uma biblioteca de visualização de dados para o Python. Esta biblioteca suporta personalização dos gráficos e é compatível com outras bibliotecas como o pandas e o numpy.

# Importação

```python
import matplotlib.pyplot as plt
```

# Propriedades de um gráfico

- **Titulo**
- **Legenda**
- **Axis** (x, y) & **Axis label**
- **label**
- **Minor tick label**
- **Major tick label**
- **Plot** (marcador)
- **Grid** (linhas tracejadas)
- **Spines** (linhas laterais)

Dados

```python
x = [1,2,3,4]
y = [2,1,6,3]
y2 = [3,5,2,6]
```

Plot

```python
plt.plot(x,y)
plt.plot(x,y2)
plt.show()
```

Linhas

```python
#plt.plot(x,y, ls = "--")
#plt.plot(x,y, ls = "-.")
#plt.plot(x,y, ls = ":")
plt.plot(x,y, ls = ":", color='brown')
plt.show()
```

Marcadores

```python
plt.plot(x,y, ls = ":", color='brown', marker='o', linewidth='2.5')
plt.plot(x,y2, ls = ":", color='blue', marker='p')
plt.show()
```

Título, Rótulos e legendas

```python
plt.plot(x,y, ls = ":", label='Valores 1', color='brown', marker='o', linewidth='2.5')
plt.plot(x,y2, ls = ":", label='Valores 2', color='blue', marker='p', linewidth='2.5')
plt.title("Meu gráfico")
plt.legend()
plt.xlabel('Tempo')
plt.ylabel('Valor')
plt.show()
```

Estilo

```python
plt.style.use('bmh')
plt.plot(x,y, ls = ":", label='Valores 1', color='brown', marker='o', linewidth='2.5')
plt.plot(x,y2, ls = ":", label='Valores 2', color='blue', marker='p', linewidth='2.5')
plt.title("Meu gráfico")
plt.legend()
plt.xlabel('Tempo')
plt.ylabel('Valor')
plt.show()
```

plt.figure()

```python
# Cria uma figura com tamanho específico
figura = plt.figure(figsize=(20, 3))

# Adiciona o primeiro subplot à figura
ax1 = figura.add_subplot(131)
ax1.plot(x, y, ls=":", label='Valores 1', color='brown', marker='o', linewidth=2.5)
ax1.set_title('Gráfico 1')
ax1.legend()

# Adiciona o segundo subplot à figura
ax2 = figura.add_subplot(132)
ax2.plot(x, y, ls="--", label='Valores 2', color='r', marker='>', linewidth=2.5)
ax2.set_title('Gráfico 2')
ax2.legend()

# Adiciona o terceiro subplot à figura
ax2 = figura.add_subplot(133)
ax2.plot(x, y2, ls="-", label='Valores 2', color='b', marker='*', linewidth=2.5)
ax2.set_title('Gráfico 3')
ax2.legend()

# Adiciona um título principal à figura
figura.suptitle('Título Principal')

# Exibe o gráfico
plt.show()
```

plt.scatter()
Gráfica de ponto

```python
plt.style.use('ggplot')
plt.scatter(x,y, marker='*', color='r', label='Scatter')
plt.title('Gráfico scatter')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

Gráfico de barra

```python
plt.style.use('ggplot')
plt.bar(x,y, color='r', label='Bar')
plt.title('Gráfico scatter')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
