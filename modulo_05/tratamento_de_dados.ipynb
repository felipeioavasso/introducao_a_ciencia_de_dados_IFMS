# Importando as bibliotecas


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import requests
```

# Importando o arquivo diretamente da internet

```python
# URL do arquivo
url = "http://leg.ufpr.br/~walmes/data/euro_football_players.txt"

# Fazendo o request para obter o conteúdo do arquivo
response = requests.get(url)

# Verifica se o request foi bem-sucedido
if response.status_code == 200:
    # Salva o conteúdo em um arquivo local
    with open("euro_football_players.txt", "w", encoding="utf-8") as file:
        file.write(response.text)
    print("Arquivo baixado e salvo com sucesso.")
else:
    print(f"Falha ao fazer o request. Código de status: {response.status_code}")
```

Lendo arquivo

```python
data = pd.read_csv('euro_football_players.txt', skiprows=27, sep='\t')
```

### Dados Faltantes

Tratar dados nulos em ciências de dados com Python pode ser feito usando diferentes técnicas, incluindo:

1. **Remoção de linhas ou colunas com dados nulos**  
   Excluir registros ou variáveis que contenham valores nulos.

2. **Substituição por valor médio, mediana ou moda**  
   Preencher os valores nulos com a média, mediana ou moda dos dados.

3. **Interpolação ou extrapolação de dados**  
   Usar métodos de interpolação ou extrapolação para estimar os valores faltantes com base em dados existentes.

4. **Previsão de valores baseada em modelos de machine learning**  
   Utilizar modelos preditivos para estimar e preencher os valores faltantes.

5. **Atribuição de valor zero ou outro valor constante**  
   Substituir os valores nulos por zero ou por um valor constante específico para o contexto.

# Substituir dados nulos por valor médio, mediana ou moda

```python
data['cm'].isnull().any()
```

```python
data['cm'].fillna(data['cm'].mean(), inplace = True)
```

```python
data.fillna({'cm': data['cm'].mean()}, inplace=True)
```

```python
#data['kg'].fillna(data['kg'].median)
data.fillna({'kg': data['kg'].median()}, inplace= True)
```

```python
data
```

```python
moda = data['goal'].mode().iloc[0]
#data['goal'].fillna(moda, inplace=True)
data.fillna({'goal': moda}, inplace=True)
```

```python
data
```

```python
data.fillna({'kg': data['kg'].median()}, inplace= True)
```

```python
data.fillna({'ass': 0}, inplace=True)
```

```python
data
```

```python
data.fillna({'red': 'Default'}, inplace=True)
data
```
